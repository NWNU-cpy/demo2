```python
file=open('D:\Course\\ 3data_set\idkp1-10.txt','r')
file1=open('D:\Course\\ 3data_set\i.txt','w')
string=file.read().strip()
ss=string.split('\n')
s=[]
for i in ss:
    i.rstrip(',')
    i.rstrip('\n')
    i.rstrip('.')
    if i.startswith('*') or i.startswith('The') or i.startswith('IDK') or i==''  :
        continue
    else:
        s.append(i[:-1])
print(s)
file1.write("\n".join(s))
file.close()
file1.close()
```

```python
import matplotlib.pyplot as plt
```

```python

fig,ax=plt.subplots()
ax1=plt.subplot(1,1,1)
for i in range(0,len(s)-1,2):
    x=list(map(int,s[i].split(',')))
    y=list(map(int,s[i+1].split(',')))
    ax1.scatter(x,y)
    plt.show()
```
